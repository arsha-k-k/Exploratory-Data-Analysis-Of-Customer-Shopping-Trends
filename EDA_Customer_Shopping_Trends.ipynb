```python
df["Gender"].value_counts().plot(kind = "bar")
plt.title("Bar plot of Gender distribution")
plt.show()
```

```python
# Plotting histograms for numerical columns
numerical_columns = ['Age', 'Purchase Amount (USD)', 'Previous Purchases']
df[numerical_columns].hist(bins=10, figsize=(10, 6), edgecolor='black')
plt.suptitle('Histograms of Numerical Columns')
plt.show()
```

```python
df["Subscription Status"].value_counts().plot(kind = "pie")
plt.title("Pie plot of Subscription Status")
plt.show()
```

```python
# Boxplot to visualize the distribution and outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[numerical_columns])
plt.title('Boxplot of Numerical Columns')
plt.show()
```

```python
df['Category'].value_counts()
```

```python
# Bar plot for categorical columns like 'Category', 'Gender', etc.
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Category')
plt.title('Countplot for Product Categories')
plt.xticks(rotation=45)
plt.show()
```

```python
season_counts = df['Season'].value_counts()
season_counts
```

```python
# Create a horizontal bar chart
plt.figure(figsize=(8, 6))
season_counts.plot(kind='barh', color=['#66b3ff', '#99ff99', '#ffcc99', '#ff9966'])
plt.title('Season Distribution')
plt.xlabel('Number of Purchases')
plt.ylabel('Season')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Size', hue='Category', palette='Set2')
plt.title('Size Distribution by Category')
plt.xlabel('Size')
plt.ylabel('Number of Purchases')
plt.show()
```

```python
payment_method_counts = df['Payment Method'].value_counts()
payment_method_counts
```

```python
# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(payment_method_counts, labels=payment_method_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99','#ffcc99','#ff9966','#c2c2f0'])
plt.title('Payment Method Distribution')
plt.axis('equal')
plt.show()
```

```python
# Heatmap for correlation of numerical columns
corr = df[numerical_columns].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Columns')
plt.show()
```
